```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt4o_llm = ChatOpenAI(model="gpt-4o", temperature=0)
response = gpt4o_llm.invoke("Hola como estas?")
response.pretty_print()
```

```python
from langchain_openai import ChatOpenAI

gpt3_llm = ChatOpenAI(model="gpt-3.5-turbo-0125", temperature=0)
response = gpt3_llm.invoke("Hola como estas?")
response.pretty_print()
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="Eres un asistente útil.")
msg_2 = HumanMessage(content="¿Cuál es la capital de Francia?")
msg_3 = AIMessage(content="La capital de Francia es París.")
msg_4 = HumanMessage(content="¿Cuál fue mi primer pregunta?")

history = [msg_1, msg_2, msg_3, msg_4]

response = gpt3_llm.invoke(history)
response.pretty_print()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

gemini_llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", temperature=0)
response = gemini_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

gemini_llm_v2 = init_chat_model("gpt-3.5-turbo-0125", temperature=0,)
response = gemini_llm_v2.invoke(history)
response.pretty_print()
```
