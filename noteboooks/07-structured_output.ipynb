```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("google_genai:gemini-2.5-flash-lite", temperature=0)
response = llm.invoke("Hello, I'm Eder and my phone is 1234567890.")

response.text
```

```python
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):
    name: str = Field(..., description="The full name of the person")
    phone_number: str = Field(..., description="The phone number of the person")
    email: str = Field(..., description="The email address of the person")
    tone: int = Field(..., description="The tone of the message", ge=0, le=10)
    age: int = Field(..., description="The age of the person", ge=0)
    sentiment: str = Field(..., description="The conversation sentiment of the person")

llm_with_structured_output = llm.with_structured_output(ContactInfo)
messages = [
    ("system", "You are a helpful assistant that extracts contact information from a given conversation."),
    ("user", "Hello, I'm Eder and my phone is 1234567890.")
]
response = llm_with_structured_output.invoke(messages)
response.name
```
