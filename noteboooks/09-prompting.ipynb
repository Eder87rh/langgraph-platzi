```python
from langchain_core.prompts import PromptTemplate

template = """\
# Instrucciones para el asistente de IA

Recuerda que hoy es {fecha_actual}.

## Rol
Eres un experto en marketing digital con 10 años de experiencia ayudando a empresas a mejorar su presencia en línea y aumentar sus ventas a través de estrategias efectivas de marketing digital.

## Tarea
Analiza el siguiente texto de un anuncio y proprciona 3 sugerencias para mejorarlo, enfocándote en atraer más clientes potenciales y aumentar la tasa de conversión.

## Texto del anuncio
{texto_anuncio}

## Formato de salida
- *Sugerencia 1*: [Tu sugerencia aquí]
- *Sugerencia 2*: [Tu sugerencia aquí]
- *Sugerencia 3*: [Tu sugerencia aquí]

"""

promptTemplate = PromptTemplate.from_template(template)
prompt = promptTemplate.format(fecha_actual="2024-06-15", texto_anuncio="Compra nuestro producto ahora y obtén un descuento del 10%!")

print(prompt)
```

```python
import datetime
today = datetime.date.today().strftime("%Y-%m-%d")

promptTemplate = PromptTemplate.from_template(partial_variables={"fecha_actual": today}, template=template)
prompt = promptTemplate.format(texto_anuncio="Compra nuestro producto ahora y obtén un descuento del 10%!")

print(prompt)
```

```python


template = """\
You are a helpful assistant, that reponds to user's messages.

{% if name %}
The user's name is {{ name }} and you can call them by that name.
{% endif -%}

The user's message is: {{ user_message }}
"""

promptTemplate = PromptTemplate.from_template(template=template, template_format="jinja2")
prompt = promptTemplate.format(user_message="Como puedo mejorar mi sitio web?", name="Carlos")

print(prompt)
```
